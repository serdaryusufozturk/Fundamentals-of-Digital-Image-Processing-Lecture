```python
import numpy as np
import matplotlib.pyplot as plt

#Format of PGM
#P5
#24 7 (width)& height
#15 (maxval)
```

## Write fonksiyonu

```python
def mypgmwrite(filename,data):
    magic_number = "P5"
    max_val = 255
    height = len(data)
    width = len(data[0])
    header = f'{magic_number}\n{str(width)} {str(height)}\n{str(max_val)}\n'
    
    with open(filename,'wb') as f:#write binary olarak dosyayı açtım, bu sayede pgm dosyasına binary yazabiliyorum.
        f.write(header.encode())#Encode ile stringleri UTF-8 veya başka bir kodlama tekniğiyle kodlar.
        for i in range(height):
            for j in range(width):
                f.write(int(data[i][j]).to_bytes(1,'big'))#1 byte 0-255 arası olduğu için 1 byte a çeviriyoruz. Maxval 255 olmasaydı bu da farklı olurdu.
```

## Read fonksiyonu

```python
def mypgmread(filename):
    with open(filename,'rb') as f:#write binary olarak dosyayı açtım, bu sayede pgm dosyasına binary yazabiliyorum.
        magic_word = f.readline().strip().decode() #Magicword. Strip metodu ile stringin başında ve sonundaki boşlukları kaldırıyoruz.
        comment = '#'
        f_content = f.readline().decode()
        while(f_content[0] == '#'):#Yorum satırlarını ayırt etmek için.
            f_content = f.readline()
#split metodu default olarak stringi boşluklardan ayırır. kelime kelime listeler.
        [width,height] = f_content.split()
        width = int(width)
        height = int(height)
        maxval = f.readline().decode()
        maxval = int(maxval)
        #print(magic_word)
        #print(f'width: {width}, height: {height}')
       # print(maxval)
        i = 0
        j = 0
        imgValue = [[0 for i in range(width)] for j in range(height)]
        byte = f.read(1)
        counter = 0
        while(counter != width*height):
            #print(f'i = {i}\nj = {j}')
            imgValue[i][j] = ord(byte)
            #print(f'byte {ord(byte)}')
            j += 1
            counter += 1
            if(j == width):
                i += 1
                j = 0
            byte = f.read(1)
    return magic_word,width,height,maxval,imgValue
```

```python
magic_word,width,height,maxval,data = mypgmread("test.pgm")
print(f'Magic word: {magic_word}\nwidth & height: {width} {height}\nmaxval: {maxval}\n{data}')
```

## Bu çalışmada şunları öğrendim.
### Magicword P5 olarak kullandığımız için dosyaya binary yazacağız. Bunu yapmak için stringleri encode() metoduyla UTF-8 olarak dosyaya kodladım. Okurken de bu string kısımları aynı şekilde decode() metoduyla decode ediyorum. Integer olan veriler ise to_bytes ile byte halinde kodlanıyor. burada maxval 255 olduğu için hepsini 1 byte olarak kodlayabiliyorum. Daha sonra bunları da integera çevirmek için ord() fonksiyonundan faydalanıyorum. Ord() fonksiyonu girilen bir karakterin unicode eşdeğerini döndürür. Daha doğrusu girilen karakterin unicode eşdeğerine gelen integer değeri döndürür. Veya argüman 8 bitlik bir dizi ise yani 1 byte ise bu baytın değerini döndürür. İşte burada ben bu amaçla yani baytın değerini döndürme amacıyla ord fonksiyonunu kullanıyorum.

### Bu çalışmada bir ara unicode ile ilgili bir hata aldım. Hatanın kaynaklanma sebebi stringlerden sonra gelen data kısmı integer olmasına rağmen veriyi alırken decode etmemden kaynaklanıyordu. Unutmayalım ki encode ve decode fonksiyonları stringlerle çalışır, integerlar ile değil.

```python
magic_word,width,height,maxval,data = mypgmread("lena.pgm")
print(f'Magic word: {magic_word}\nwidth & height: {width} {height}\nmaxval: {maxval}\n{data}')
#data verisinde bütün satırlar ve sütunlar mevcut.Bu satırlardan 30 ve 50 arasını aşağıdaki döngüyle 0 yapıyorum. Daha sonra
#bu yeni datayı myLena.pgm adındaki dosyaya yazdırıyorum.
for i in range(0,256,1):
    for j in range(0,256,1):
        data[i][j] = abs(data[i][j]-256)
mypgmwrite("negative.pgm",data)
```
